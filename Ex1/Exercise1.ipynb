# Число pi Методом Монте-Карло

## Площадь квадрата $S = (2R)^2$
## Площадь круга $C = πR^2$
## $C/S = (π R^2)/(4R^2) = π /4$
## $π = 4 * C/S$

```python
import random
import matplotlib.pyplot as plt
```

```python
num_points = 100000
```

```python
points = []
hits = 0
for _ in range (num_points):
    x, y = random.random(), random.random()
    if x*x + y*y < 1.0:
        hits += 1
        points.append((x,y, "black"))
    else:
        points.append((x,y, "crimson"))
```

```python
x, y, colors = zip(*points)

fig, ax = plt.subplots()
fig.set_size_inches(7.0, 7.0)

ax.scatter(x, y, c=colors)
```

```python
from matplotlib import pyplot
import random
from ipywidgets import interact, widgets

%matplotlib inline


def throw_darts(num_points):
    points = []
    hits = 0
    for _ in range(num_points):
        x, y = random.random(), random.random()
        if x*x + y*y < 1.0:
            hits += 1
            points.append((x, y, "darkorchid"))
        else:
            points.append((x, y, "lawngreen"))
    fraction = hits / num_points
    pi = 4 * fraction
    return pi, points


def create_plot(points):
    x, y, colors = zip(*points)
    pyplot.scatter(x, y, c=colors)


def experiment(num_points):
    pi, points = throw_darts(num_points)
    create_plot(points)
    print("approximation:", pi)
```

```python
interact(experiment, num_points=widgets.IntSlider(min=100, max=10000, step=100, value=10000))
```
